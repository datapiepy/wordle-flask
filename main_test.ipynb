```python
import random
guess = input("Podaj słowo:").strip().lower()

def load_words(path: str) -> list[str]:
    with open(path, "r", encoding="utf-8") as f:
        words = [line.strip().lower() for line in f]
    words = [w for w in words if len(w) == 5 and w.isalpha()]
    return words

words = load_words("wordlist_fives.txt")
words_set = set(words)
secret = random.choice(words)
secret

def is_valid(guess, words_set):
    return len(guess) == 5 and guess.isalpha() and guess in words_set

def score(guess, secret):
    word_guessed = list(guess)
    word_secret = list(secret)
    result = []

    for idx in range(5):
        if word_guessed[idx] == word_secret[idx]:
            result.append("G")
        elif word_guessed[idx] in word_secret:
            result.append("Y")
        else:                
            result.append("B")
    return "".join(result)


def play(guess, secret):
    for attempt in range(6):
        guess = input(f"Próba {attempt+1}/6, podaj słowo:").strip().lower()

        while not is_valid(guess, words_set):
            print( "Błędne słowo - nie ma 5 znaków")
            guess = input("Podaj słowo:").strip().lower()
        
        feedback = score(guess, secret)
        print(feedback)
    
        if feedback == "GGGGG":
            print("Wygrana")
            break
    else:
        print(f"Koniec prób! Hasłow to: {secret}")
    
```

```python
play(guess, secret)
```
